```python
import pandas as pd
df = pd.read_csv('dataset.csv', encoding="utf-8-sig")
# col_list = ["Rank", "Presence_of_Keyword_in_Title", "Title_Starts_with_Keyword", "Presence_Of_Keyword_In_Url", "Keyword_in_Description", "Presence_of_Internal_Link", "Presence_of_External_Links"]
# df = pd.read_csv("file10.csv", usecols=col_list)
df.shape
```

```python
import math
import numpy as np 
df['class'] = (df['Rank'] / 10).apply(np.ceil)
df['class'] = ((df['class'] - 1)/10).apply(np.ceil)
# print(df['class'].to_string())
df['class'].unique()
```

```python
df = df.drop(columns=['Keyword', 'Title_Of_Page', 'URL', 'Domain_Name'])
```

```python
df.sample(30)
# print( df['class'].to_string())
```

```python
X = df.drop(columns=['Rank', 'class'])
y = df['class']
y.unique()
X
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train)
X_train = sc.transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Importing the Keras libraries and packages
# !pip install keras
# !pip install tensorflow

import keras
from keras.models import Sequential
from keras.layers import Dense


# Initialising the ANN
classifier = Sequential()
```

```python
# Input layer
classifier.add(Dense(units = 24, kernel_initializer = 'uniform', activation = 'relu'))
# Adding the output layer
classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
```

```python
# Compiling the ANN
#adam is implemenation version of stocahastic gradient descent.
from keras.optimizers import Adam

classifier.compile(optimizer = 'Adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
# Fitting the ANN to the Training set
classifier.fit(X_train, y_train, batch_size = 10, epochs = 1000)
```

```python
# Part 3 - Making predictions and evaluating the model

# Predicting the Test set results
y_pred = classifier.predict(X_test)

y_pred = (y_pred > 0.5)
print(y_pred)
```

```python
# Evaluate the model on test set
classifier.evaluate(X_test, y_test)
print(y_test)
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```
